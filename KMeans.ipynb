```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

sns.set()
from sklearn.cluster import KMeans as kn
```

```python
data = pd.read_csv("winequality.csv")
data.head()
```

```python
#plotting data
plt.scatter(data['fixed acidity'],data['quality'])
plt.title("Fixed acidity VS quality")
plt.show()

plt.scatter(data['volatile acidity'],data['quality'])
plt.title("volatile acidity VS quality")
plt.show()

plt.scatter(data['citric acid'],data['quality'])
plt.title("citric acid VS quality")
plt.show()

plt.scatter(data['residual sugar'],data['quality'])
plt.title("residual sugar VS quality")
plt.show()

plt.scatter(data['chlorides'],data['quality'])
plt.title("chlorides VS quality")
plt.show()

plt.scatter(data['free sulfur dioxide'],data['quality'])
plt.title("free sulfur dioxide VS quality")
plt.show()

plt.scatter(data['total sulfur dioxide'],data['quality'])
plt.title("total sulfur dioxide VS quality")
plt.show()

plt.scatter(data['density'],data['quality'])
plt.title("density VS quality")
plt.show()

plt.scatter(data['pH'],data['quality'])
plt.title("ph VS quality")
plt.show()

plt.scatter(data['sulphates'],data['quality'])
plt.title("sulphates VS quality")
plt.show()

plt.scatter(data['alcohol'],data['quality'])
plt.title("alcohol VS quality")
plt.show()
```

```python
x = data.iloc[:,0:11]
x
```

```python
#centroids point
kmeans = kn(2)
kmeans
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['clusters'] = identified_clusters
plt.scatter(data_with_clusters['fixed acidity'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("Fixed acidity VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['volatile acidity'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("volatile acidity VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['citric acid'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("citric acid VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['residual sugar'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("residual VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['chlorides'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("chlorides VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['free sulfur dioxide'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("free sulfur dioxide VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['total sulfur dioxide'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("total sulfur dioxide VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['density'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("density VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['pH'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("ph VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['sulphates'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("sulphates VS Quality with clusters")
plt.show()

plt.scatter(data_with_clusters['alcohol'],data_with_clusters['quality'],
            c=data_with_clusters['clusters'],cmap='rainbow')
plt.title("alcohol VS Quality with clusters")
plt.show()
```
