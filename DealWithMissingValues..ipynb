Attribute Description:- Input variables (based on physicochemical tests): 1 - fixed acidity 2 - volatile acidity 3 - citric acid 4 - residual sugar 5 - chlorides 6 - free sulfur dioxide 7 - total sulfur dioxide 8 - density 9 - pH 10 - sulphates 11 - alcohol Output variable (based on sensory data): 12 - quality (score between 0 and 10)

Aim :- Implement the techniques to deal with missing values.

WE ARE FILLING THE NULL DATA WITH THE HELP OF MEAN,MEDIAN,AND MODE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
%matplotlib inline

df = pd.read_csv('winequality.csv')

df.head(20)
```

```python
#check if there is any null value in dataset
df.isnull().any()
```

```python
#getting mean,median,mode of these data 
lstFA = df['fixed acidity']
lstVA = df['volatile acidity']
lstCA = df['citric acid']
lstRS = df['residual sugar']
lstFSD = df['free sulfur dioxide']
lstTSD = df['total sulfur dioxide']
lstDEN = df['density']
lstSulp = df['sulphates']
lstAl = df['alcohol']

#we cant use mean because it has some null values
medianFA = st.median(lstFA)
modeFA = st.mode(lstFA)

medianVA = st.median(lstVA)
modeVA = st.mode(lstVA)

medianCA = st.median(lstCA)
modeCA = st.mode(lstCA)

medianRS = st.median(lstRS)
#modeRS = st.mode(lstRS) #cant use because it has more than one mode

medianFSD = st.median(lstFSD)
#modeFSD = st.mode(lstFSD) #cant use because it has more than one mode

medianTSD = st.median(lstTSD)
#modeTSD = st.mode(lstTSD) #cant use becuase it has more than one mode

medianDEN = st.median(lstDEN)
modeDEN = st.mode(lstDEN)

medianSP = st.median(lstSulp)
modeSP = st.mode(lstSulp)

medianAl = st.median(lstAl) # cant use  because it gives an NaN value
modeAl = st.mode(lstAl)

print("FA mean: "+str(medianFA)+" FA mode: "+str(modeFA))
print("VA mean: "+str(medianVA)+" VA mode: "+str(modeVA))
print("CA mean: "+str(medianCA)+" CA mode: "+str(modeCA))
print("RS mean: "+str(medianRS))#+"RS mode: "+str(modeRS))
print("FSD mean: "+str(medianFSD))#+"FSD mode: "+str(modeFSD))
print("TSD mean: "+str(medianTSD))#+"TSD mode: "+str(modeTSD))
print("DEN mean: "+str(medianDEN)+" DEN mode: "+str(modeDEN))
print("SP mean: "+str(medianSP)+" SP mode: "+str(modeSP))
print("Al mean: "+str(medianAl)+" Al mode: "+str(modeAl))
```

```python
def fillNull():
    new_dataSet = df
    if df['fixed acidity'].isnull().any():
        new_dataSet['fixed acidity'] = df['fixed acidity'].fillna(medianFA)
    if df['volatile acidity'].isnull().any():
        new_dataSet['volatile acidity'] = df['volatile acidity'].fillna(medianVA)
    if df['citric acid'].isnull().any():
        new_dataSet['citric acid'] = df['citric acid'].fillna(medianCA)
    if df['residual sugar'].isnull().any():
        new_dataSet['residual sugar'] = df['residual sugar'].fillna(medianRS)
    if df['free sulfur dioxide'].isnull().any():
        new_dataSet['free sulfur dioxide'] = df['free sulfur dioxide'].fillna(medianFSD)
    if df['total sulfur dioxide'].isnull().any():
        new_dataSet['total sulfur dioxide'] = df['total sulfur dioxide'].fillna(medianTSD)
    if df['density'].isnull().any():
        new_dataSet['density'] = df['density'].fillna(medianDEN)
    if df['sulphates'].isnull().any():
        new_dataSet['sulphates'] = df['sulphates'].fillna(medianSP)
    if df['alcohol'].isnull().any():
        new_dataSet['alcohol'] = df['alcohol'].fillna(modeAl)
    return new_dataSet
```

```python
new_dataSet = fillNull()
new_dataSet.isnull().any()
```

```python
sns.heatmap(new_dataSet.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```
